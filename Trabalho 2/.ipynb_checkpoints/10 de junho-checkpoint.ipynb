 - Agora determinaremos os pixels que queremos somar

```python
y = np.arange(new_data.shape[0])
x = np.arange(new_data.shape[1])

x, y = np.meshgrid(x, y)
```

```python
abertura = 10

mask = np.sqrt((x - x0)**2 + (y - y0)**2) < abertura

image_copy = np.copy(new_data.astype('float64')) # Puramente para ilustração
image_copy[~mask] = np.nan  # Puramente para ilustração
```

```python
plt.figure(figsize=[5, 5])
plt.imshow(new_data, cmap='gray', 
          #vmin=im_data.mean()-100, 
          #vmax=im_data.mean()+100
          )

plt.imshow(image_copy, cmap='Reds', 
          #vmin=im_data.mean()-100, 
          #vmax=im_data.mean()+100
          )
plt.plot(x0, y0, 'go')
plt.show()
```

```python
fluxo = np.sum(new_data[mask])
print(fluxo)

npix = len(new_data[mask])
print(npix)
```

```python
fluxos = np.array([])
abertura = np.arange(1,30)

for i in abertura:
    mask = np.sqrt((x - x0)**2 + (y - y0)**2) < i
    fluxos = np.append(fluxos, np.sum(new_data[mask]))
```

```python
plt.plot(fluxos, abertura, 'k.-')
plt.ylabel('Abertura Circular [px]', fontsize=15)
plt.xlabel('Fuxo [contagens]', fontsize=15)
```

```python
snr = fluxo/np.sqrt(fluxo)

print('S/N:', snr)
```

 - determinando fundo de céu

```python
outer_ring = 10
inner_ring = 7

mask_sky = ((np.sqrt((x - x0)**2 + (y - y0)**2) < outer_ring) 
           *(np.sqrt((x - x0)**2 + (y - y0)**2) > inner_ring))


image_sky = np.copy(new_data.astype('float64')) # Puramente para ilustração
image_sky[~mask_sky] = np.nan  # Puramente para ilustração
```

```python
plt.figure(figsize=[5, 5])
plt.imshow(new_data, cmap='gray')
plt.imshow(image_sky, cmap='Reds', alpha=1.0)
#pl.plot(x0, y0, 'ro')
plt.show()
```

```python
flux_sky = np.sum(new_data[mask_sky])
print('Flux do céu:', flux_sky)

npix_sky = len(new_data[mask_sky])
print('Número de pixeis:', npix_sky)
print('\n')
print('Distribuição do Fundo do céu:' + 
      '{:.2f} +/- {:.2f}'.format(new_data[mask_sky].mean(),
                                 new_data[mask_sky].std()))
```

```python
fluxo_correto = fluxo - npix*new_data[mask_sky].mean()

snr = (fluxo_correto/np.sqrt((fluxo_correto + 
       npix*new_data[mask_sky].mean())))

print('S/N:', snr) #Realista!
```

```python
fluxos = np.array([])
npixs = np.array([])

abertura = np.arange(1,15)
for i in abertura:
    mask = np.sqrt((x - x0)**2 + (y - y0)**2) < i
    fluxos = np.append(fluxos, np.sum(new_data[mask]))
    npixs  = np.append(npixs, len(new_data[mask]))

    
fluxos_new = fluxos - npixs*new_data[mask_sky].mean()

snr = fluxos_new/np.sqrt(fluxos_new + 
                         npixs*new_data[mask_sky].mean())
```

```python
plt.plot(abertura, snr, 'k.-')
plt.xlabel('Abertura Circular [px]', fontsize=15)
plt.ylabel('S/N', fontsize=15)
plt.axvline(5, color='r')
plt.axhline(587.83, color = 'g')
plt.show()
```

```python
np.where(snr == snr.max())
```
