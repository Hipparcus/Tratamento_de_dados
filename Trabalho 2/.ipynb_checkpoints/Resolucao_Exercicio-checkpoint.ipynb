### Começamos fazendo um corte em torno da região de interesse

```python
def image_cut(image, x0, y0, dx, dy):
    im_cut = np.copy(image[int(y0-dy/2):int(y0+dy/2), int(x0-dx/2):int(x0+dx/2)])
    return im_cut
```

```python
x0=254
y0=247
dx=30
dy=30

image = image_cut(image=image_full,x0=x0,y0=y0, dx=dx, dy=dy)
```

```python
pl.figure(figsize=[10, 10])
pl.imshow(image, cmap='gray', vmin=image.mean()-1000, vmax=image.mean()+1000)
pl.show()
```

### Vamos determinar o fotocentro do alvo

```python
def momento(data):
    '''
    Função que devolve o fotocentro da imagem utilizando o método do momento.
    '''
    I_i = np.sum(data, axis=0)
    I_j = np.sum(data, axis=1)
    Ii_mean = np.sum(I_i)/len(I_i)
    Ij_mean = np.sum(I_j)/len(I_j)

    x_i = np.arange(data.shape[1])
    y_j = np.arange(data.shape[0])
    mask_i = (I_i - Ii_mean) > 0
    mask_j = (I_j - Ij_mean) > 0

    xc = np.sum((I_i - Ii_mean)[mask_i]*x_i[mask_i])/np.sum((I_i - Ii_mean)[mask_i])
    yc = np.sum((I_j - Ij_mean)[mask_j]*y_j[mask_j])/np.sum((I_j - Ij_mean)[mask_j])
    return xc, yc
```

```python
xc, yc = momento(image)
print(xc, yc)
```

```python
pl.figure(figsize=[10, 10])
pl.imshow(image, cmap='gray', vmin=image.mean()-1000, vmax=image.mean()+1000)

pl.axhline(yc, color='blue')
pl.axvline(xc, color='blue')

pl.show()
```

```python
# Centragem na imagem original:
xc_alvo = xc + int(x0-dx/2)
yc_alvo = yc + int(y0-dy/2)

print('Center x: {:.3f}'.format(xc_alvo))
print('Center y: {:.3f}'.format(yc_alvo))
```

### Vamos determinar os pixeis que queremos somar (máscara)

```python
# Vamos criar umas variaveis que irão nos ajudar
y = np.arange(image.shape[0])
x = np.arange(image.shape[1])

x, y = np.meshgrid(x, y)
```

```python
outer_ring = 14
inner_ring = 10

abertura = 6

mask = np.sqrt((x - xc)**2 + (y - yc)**2) < abertura

mask_sky = ((np.sqrt((x - xc)**2 + (y - yc)**2) < outer_ring) * 
            (np.sqrt((x - xc)**2 + (y - yc)**2) > inner_ring))

image_sky = np.copy(image) # Puramente para ilustração
image_copy = np.copy(image) # Puramente para ilustração
image_sky[~mask_sky] = np.nan  # Puramente para ilustração
image_copy[~mask] = np.nan  # Puramente para ilustração



pl.figure(figsize=[10, 10])
pl.imshow(image, cmap='gray', vmin=image.mean()-1000, vmax=image.mean()+1000)
pl.imshow(image_sky, cmap='Reds', alpha=1.0, vmin=image.mean()-1000, vmax=image.mean()+1000)
pl.imshow(image_copy, cmap='Blues', alpha=1.0, vmin=image.mean()-1000, vmax=image.mean()+1000)

pl.show()
```

```python
def sigma_clip(value, sigma):
    return value[np.absolute(value - value.mean()) < sigma*value.std(ddof=1)]

def photometry(image, xc, yc, abertura_min, abertura_max, sky_min, sky_max, n_sigma_clip=3):

    mask_sky = ((np.sqrt((x - xc)**2 + (y - yc)**2) < sky_max) * 
                (np.sqrt((x - xc)**2 + (y - yc)**2) > sky_min))

    pl.figure(figsize=[8,5])
    pl.hist(image[mask_sky], 20)
    pl.axvline(image[mask_sky].mean(), color='red')
    pl.axvline(image[mask_sky].mean() + 3*image[mask_sky].std(), color='red', ls='--')
    pl.axvline(image[mask_sky].mean() - 3*image[mask_sky].std(), color='red', ls='--')
    pl.show()
    print('Distribuição do Fundo do céu: {:.2f} +/- {:.2f}'.format(image[mask_sky].mean(), image[mask_sky].std()))

    sky_flux_cut = sigma_clip(value=image[mask_sky].ravel(), sigma=n_sigma_clip)
    print('Distribuição do Fundo do céu: {:.2f} +/- {:.2f}'.format(sky_flux_cut.mean(), sky_flux_cut.std()))

    pl.figure(figsize=[8,5])
    pl.hist(sky_flux_cut, 20)
    pl.axvline(sky_flux_cut.mean(), color='blue')
    pl.axvline(sky_flux_cut.mean() + 3*sky_flux_cut.std(), color='blue', ls='--')
    pl.axvline(sky_flux_cut.mean() - 3*sky_flux_cut.std(), color='blue', ls='--')
    pl.show()
    
    mean_sky = sky_flux_cut.mean()
    
    fluxos = np.array([])
    npixs = np.array([])

    abertura = np.arange(abertura_min, abertura_max, 1)
    for i in abertura:
        mask = np.sqrt((x - xc)**2 + (y - yc)**2) < i
        fluxos = np.append(fluxos, np.sum(image[mask]))
        npixs  = np.append(npixs, len(image[mask]))

    fluxos_new = fluxos - npixs*mean_sky
    snr = fluxos_new/np.sqrt(fluxos_new + npixs*mean_sky)

    print('Melhor abertura: {:02d} px'.format(abertura[snr.argmax()]))

    pl.plot(abertura, snr, 'k.-')
    pl.xlabel('Abertura Circular [px]', fontsize=15)
    pl.ylabel('S/N', fontsize=15)
    pl.axvline(abertura[snr.argmax()], color='k', ls='--')
    pl.show()    
    return fluxos_new[snr.argmax()], snr.max(), abertura[snr.argmax()], npixs[snr.argmax()]
```

```python
param = photometry(image=image, xc=xc, yc=yc, 
                   abertura_min=1, abertura_max=10, 
                   sky_min=10, sky_max=14, n_sigma_clip=3)

flux_alvo, snr_alvo, abertura_alvo, npix_alvo = param
```

```python
# Visualizando a fotometria
outer_ring = 14
inner_ring = 10

abertura = abertura_alvo

mask = np.sqrt((x - xc)**2 + (y - yc)**2) < abertura

mask_sky = ((np.sqrt((x - xc)**2 + (y - yc)**2) < outer_ring) * 
            (np.sqrt((x - xc)**2 + (y - yc)**2) > inner_ring))

image_sky = np.copy(image) # Puramente para ilustração
image_copy = np.copy(image) # Puramente para ilustração
image_sky[~mask_sky] = np.nan  # Puramente para ilustração
image_copy[~mask] = np.nan  # Puramente para ilustração

pl.figure(figsize=[10, 10])
pl.imshow(image, cmap='gray', vmin=image.mean()-1000, vmax=image.mean()+1000)
pl.imshow(image_sky, cmap='Reds', alpha=1.0, vmin=image.mean()-1000, vmax=image.mean()+1000)
pl.imshow(image_copy, cmap='Blues', alpha=1.0, vmin=image.mean()-1000, vmax=image.mean()+1000)

pl.show()
```
